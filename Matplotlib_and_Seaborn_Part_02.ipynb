### Plot the bar chart from the NaN tabular data, and also print values on each bar

```python
# The first argument to the function below contains the x-values (column names), the second argument the y-values (our counts).

# na_count.index.values will get firt the index e.g. Store which will return the base-index and then using .values it will get
# only values from that base-index which will be a numpy array
sb.barplot(x=na_count.index.values,y=na_count.values);
plt.xticks(rotation=90);


# Logic to print values on each bar

for i in range(na_count.shape[0]):
    count = na_count[i]
    
#     Below i is used x-cordinate and count+80 is used for y-cordinate ha=horizontalalighnment as so on.
    plt.text(i,count+90,s=count,ha='center',fontsize=10,color='red',fontfamily='cursive')

```

### **TO DO Task 2** 
1. Each Pokémon species has either `type_1`, `type_2` or both `types` that play a part in its offensive and defensive capabilities. The code below creates a new dataframe `pkmn_types` that club the rows of both `type_1` and `type_2`, so that the resulting dataframe has **new** column, `type_level`. 

**Display, how frequent is each type?**



The function below will do the following in the pokemon dataframe *out of place*:
1. Select the 'id', and 'species' columns from pokemon. 
2. Remove the 'type_1', 'type_2' columns from pokemon
3. Add a new column 'type_level' that can have a value either 'type_1' or 'type_2'
4. Add another column 'type' that will contain the actual value contained in the 'type_1', 'type_2' columns. For example, the first row in the pokemon dataframe having `id=1`	and `species=bulbasaur` will now occur twice in the resulting dataframe after the `melt()` operation. The first occurrence will have `type=grass`, whereas, the second occurrence will have `type=poison`.

```python
pkmn_types = pokemon.melt(id_vars = ['id','species'], 
                          value_vars = ['type_1', 'type_2'], 
                          var_name = 'type_level', value_name = 'type').dropna()
pkmn_types.head()
```

```python
# YOUR CODE HERE
# Display the frequency of each type.
pkmn_types['type_level'].value_counts()
```

2. Your task is to use this dataframe to create a _relative frequency_ plot of the proportion of Pokémon with each type, _sorted_ from most frequent to least. **Hint**: The sum across bars should be greater than 100%, since many Pokémon have two types. Keep this in mind when considering a denominator to compute relative frequencies.

```python
type_count = pkmn_types['type'].value_counts()
type_order = type_count.index
```

```python
# YOUR CODE HERE
pkmn_types['type'].value_counts()

n_pokemon = pkmn_types['type'].value_counts().sum()
max_type_count = type_count[0]
max_prop = max_type_count/ n_pokemon

# establish tick location and create plot
tick_props = np.arange(0,max_prop,0.02)
tick_names = ["{:0.2f}".format(i) for i in tick_props]
tick_names

base_color = sb.color_palette()[0]
sb.countplot(data=pkmn_types,y='type',color=base_color,order=type_order);
```

<div class='alert alert-success';><h2 style='color:white';>Histograms</h2>
<h6>A histogram is used to plot the distribution of a numeric variable. It's the quantitative version of the bar chart. However, rather than plot one bar for each unique numeric value, values are grouped into continuous bins, and one bar for each bin is plotted to depict the number. You can use either Matplotlib or Seaborn to plot the histograms. There is a mild variation in the specifics, such as plotting gaussian-estimation line along with bars in Seabron's distplot(), and the arguments that you can use in either case.</h6>
    </div>

```python
# Let's take the following 'speed' column to create a histogram chart
pokemon.head()
```

```python
# sb.histplot(pokemon,x='speed'); # Seaborn is more attractive than matplotlib however we will try now matplotlib 

# with the use of minimum and maximum of speed column we make an array of bins 5 e.g. (0-4 , 4-9 , 9-14 and so on.)
minimum , maximum = pokemon['speed'].min() , pokemon['speed'].max()
bins = np.arange(0,maximum+5,minimum)
hist_bin = len(bins)

plt.hist(data = pokemon, x = 'speed',bins=bins,edgecolor='black');
plt.xlabel(f'A historam having {hist_bin}',fontdict={'fontsize':18,'fontweight':'bold'});


```

```python
# Alternative approach to the above matplotlib histogram is seaborn.distplot()
sb.histplot(pokemon,x='speed',kde=True);
```
